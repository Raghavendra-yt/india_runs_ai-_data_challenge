## Define Job Description and Preprocess Candidate Data

```python
from sentence_transformers import SentenceTransformer
import numpy as np

# --- 1. Define a Sample Job Description (Placeholder) ---
# In a real scenario, this would be loaded from a file (e.g., job_description.docx)
job_description_text = """
We are looking for a highly skilled Software Engineer with expertise in Python, Machine Learning, and Cloud platforms (AWS/Azure/GCP). The ideal candidate will have 5+ years of experience, a strong background in data structures and algorithms, and experience in building scalable backend systems. Experience with natural language processing (NLP) or large language models (LLMs) is a significant plus.
"""
print("Sample Job Description defined.")

# --- 2. Initialize Embedding Model ---
# Using a pre-trained sentence-transformer model for semantic embeddings
model_name = 'sentence-transformers/all-MiniLM-L6-v2'
print(f"\nInitializing Sentence Transformer model: {model_name}...")
embedding_model = SentenceTransformer(model_name)
print("Embedding model loaded.")

# --- 3. Preprocess Candidate Data for Embeddings ---
# Combine relevant text fields from candidate profiles into a single string for embedding
def get_candidate_text_for_embedding(candidate):
    profile_summary = candidate['profile']['summary'] if 'profile' in candidate and 'summary' in candidate['profile'] else ''
    career_history_desc = " ".join([job['description'] for job in candidate['career_history'] if 'description' in job])
    skills_names = " ".join([skill['name'] for skill in candidate['skills'] if 'name' in skill])
    education_details = " ".join([f"{edu['degree']} in {edu['field_of_study']}" for edu in candidate['education'] if 'degree' in edu and 'field_of_study' in edu])

    # Combine all relevant text into a single string
    combined_text = f"{profile_summary} {career_history_desc} {skills_names} {education_details}"
    return combined_text.strip()

print("\nExtracting text from candidate profiles for embedding...")
# Apply the function to each row of the DataFrame
# Ensure df_candidates is not empty before proceeding
if not df_candidates.empty:
    df_candidates['combined_text_for_embedding'] = df_candidates.apply(get_candidate_text_for_embedding, axis=1)
    print(f"Combined text for embedding generated for {len(df_candidates)} candidates.")
    print("\nSample combined text for CAND_0000001:")
    print(df_candidates.loc[0, 'combined_text_for_embedding'][:500] + "...")
else:
    print("No candidates to process. df_candidates is empty.")
```

## Generate Embeddings and Calculate Semantic Similarity

```python
# --- 4. Generate Embeddings for JD and Candidates ---
print("\nGenerating embedding for Job Description...")
jd_embedding = embedding_model.encode(job_description_text)
print("JD embedding generated.")

if not df_candidates.empty:
    print("\nGenerating embeddings for all candidate profiles (this may take a few minutes for 100k candidates)...")
    candidate_embeddings = embedding_model.encode(
        df_candidates['combined_text_for_embedding'].tolist(),
        batch_size=256, # Optimized batch size for CPU
        show_progress_bar=True,
        convert_to_numpy=True
    )
    print("Candidate embeddings generated.")

    # --- 5. Calculate Semantic Similarity ---
    # Using cosine similarity between JD embedding and candidate embeddings
    from sklearn.metrics.pairwise import cosine_similarity

    print("\nCalculating semantic similarity scores...")
    # Reshape jd_embedding for cosine_similarity function (expects 2D array)
    jd_embedding_reshaped = jd_embedding.reshape(1, -1)

    # Calculate cosine similarity between the JD and all candidate embeddings
    semantic_scores = cosine_similarity(jd_embedding_reshaped, candidate_embeddings)[0]

    df_candidates['semantic_score'] = semantic_scores

    print("Semantic scores calculated and added to DataFrame.")
    print("\nCandidates with highest semantic similarity:")
    display(df_candidates[['candidate_id', 'semantic_score']].sort_values(by='semantic_score', ascending=False).head())
else:
    print("Cannot generate embeddings or calculate similarity: df_candidates is empty.")
```

## Advanced Filtering: Prioritizing Experience and Skills

```python
import re

# --- 6. Extract Years of Experience ---
# Ensure the 'profile' column exists and is a dictionary
if 'profile' in df_candidates.columns:
    df_candidates['years_of_experience'] = df_candidates['profile'].apply(lambda x: x.get('years_of_experience', 0) if isinstance(x, dict) else 0)
    print("Years of experience extracted.")
else:
    df_candidates['years_of_experience'] = 0
    print("Warning: 'profile' column not found, years of experience set to 0 for all candidates.")

# --- 7. Extract Required Skills from Job Description ---
# Using regex to find common technical skills mentioned in the JD
def extract_required_skills_from_jd(jd_text):
    # This is a simplified extraction. A more robust solution might use an NLP library
    # or explicit skill lists from a structured JD.
    keywords = ["Python", "Machine Learning", "Cloud", "AWS", "Azure", "GCP",
                "Data Structures", "Algorithms", "Backend Systems", "NLP", "LLMs"]
    required_skills = []
    for keyword in keywords:
        # FIX: Changed r'\\b' to r'\b' for correct word boundary matching
        if re.search(r'\b' + re.escape(keyword) + r'\b', jd_text, re.IGNORECASE):
            required_skills.append(keyword.lower())
    return list(set(required_skills))

required_skills = extract_required_skills_from_jd(job_description_text)
print(f"\nRequired skills from JD: {required_skills}")

# --- 8. Calculate Hard Skill Match Score ---
def calculate_skill_match_score(candidate_skills, required_skills):
    if not required_skills:
        return 0.0 # No required skills, so no match score

    candidate_skill_names = set([skill['name'].lower() for skill in candidate_skills if 'name' in skill])
    matched_skills = candidate_skill_names.intersection(required_skills)
    return len(matched_skills) / len(required_skills)

print("\nCalculating skill match scores...")
df_candidates['skill_match_score'] = df_candidates['skills'].apply(lambda x: calculate_skill_match_score(x, required_skills))
print("Skill match scores calculated.")

# --- 9. Implement Combined Scoring Logic (Prioritizing Experience) ----
def custom_ranking_score(row, jd_min_experience=5, exp_boost_threshold=10):
    semantic_score = row['semantic_score']
    years_exp = row['years_of_experience']
    skill_match = row['skill_match_score']

    # Base score combines semantic and skill match
    score = (0.6 * semantic_score) + (0.4 * skill_match)

    # Apply experience boost
    # Significant boost for experienced candidates (10-15 years)
    if years_exp >= exp_boost_threshold:
        score *= (1 + (years_exp - exp_boost_threshold + 1) * 0.05) # +5% per year over threshold
    elif years_exp >= jd_min_experience: # Moderate boost for meeting minimum experience
        score *= (1 + years_exp * 0.02) # +2% per year for meeting JD min
    elif years_exp < jd_min_experience and skill_match == 1.0: # Small boost for freshers with perfect skills
        score *= 1.05 # 5% boost for perfect skills even if less experienced
    else:
        score *= (1 - (jd_min_experience - years_exp) * 0.05) # Penalty for significantly less experience

    return max(0, score) * 10 # Scale score from 0-1 to 0-10

print("\nApplying custom ranking logic...")
df_candidates['final_ranking_score'] = df_candidates.apply(custom_ranking_score, axis=1)
print("Custom ranking scores calculated.")

print("\nTop 10 Candidates after custom ranking:")
display(df_candidates[['candidate_id', 'years_of_experience', 'skill_match_score', 'semantic_score', 'final_ranking_score']].sort_values(by='final_ranking_score', ascending=False).head(10))
```

## Advanced Filtering: Prioritizing Experience and Skills

```python
import re

# --- 6. Extract Years of Experience ---
# Ensure the 'profile' column exists and is a dictionary
if 'profile' in df_candidates.columns:
    df_candidates['years_of_experience'] = df_candidates['profile'].apply(lambda x: x.get('years_of_experience', 0) if isinstance(x, dict) else 0)
    print("Years of experience extracted.")
else:
    df_candidates['years_of_experience'] = 0
    print("Warning: 'profile' column not found, years of experience set to 0 for all candidates.")

# --- 7. Extract Required Skills from Job Description ---
# Using regex to find common technical skills mentioned in the JD
def extract_required_skills_from_jd(jd_text):
    # This is a simplified extraction. A more robust solution might use an NLP library
    # or explicit skill lists from a structured JD.
    keywords = ["Python", "Machine Learning", "Cloud", "AWS", "Azure", "GCP",
                "Data Structures", "Algorithms", "Backend Systems", "NLP", "LLMs"]
    required_skills = []
    for keyword in keywords:
        if re.search(r'\b' + re.escape(keyword) + r'\b', jd_text, re.IGNORECASE):
            required_skills.append(keyword.lower())
    return list(set(required_skills))

required_skills = extract_required_skills_from_jd(job_description_text)
print(f"\nRequired skills from JD: {required_skills}")

# --- 8. Calculate Hard Skill Match Score ---
def calculate_skill_match_score(candidate_skills, required_skills):
    if not required_skills:
        return 0.0 # No required skills, so no match score

    candidate_skill_names = set([skill['name'].lower() for skill in candidate_skills if 'name' in skill])
    matched_skills = candidate_skill_names.intersection(required_skills)
    return len(matched_skills) / len(required_skills)

print("\nCalculating skill match scores...")
df_candidates['skill_match_score'] = df_candidates['skills'].apply(lambda x: calculate_skill_match_score(x, required_skills))
print("Skill match scores calculated.")

# --- 9. Implement Combined Scoring Logic (Prioritizing Experience) ---
def custom_ranking_score(row, jd_min_experience=5, exp_boost_threshold=10):
    semantic_score = row['semantic_score']
    years_exp = row['years_of_experience']
    skill_match = row['skill_match_score']

    # Base score combines semantic and skill match
    score = (0.6 * semantic_score) + (0.4 * skill_match)

    # Apply experience boost
    # Significant boost for experienced candidates (10-15 years)
    if years_exp >= exp_boost_threshold:
        score *= (1 + (years_exp - exp_boost_threshold + 1) * 0.05) # +5% per year over threshold
    elif years_exp >= jd_min_experience: # Moderate boost for meeting minimum experience
        score *= (1 + years_exp * 0.02) # +2% per year for meeting JD min
    elif years_exp < jd_min_experience and skill_match == 1.0: # Small boost for freshers with perfect skills
        score *= 1.05 # 5% boost for perfect skills even if less experienced
    else:
        score *= (1 - (jd_min_experience - years_exp) * 0.05) # Penalty for significantly less experience

    return max(0, score) # Ensure score is not negative

print("\nApplying custom ranking logic...")
df_candidates['final_ranking_score'] = df_candidates.apply(custom_ranking_score, axis=1)
print("Custom ranking scores calculated.")

print("\nTop 10 Candidates after custom ranking:")
display(df_candidates[['candidate_id', 'years_of_experience', 'skill_match_score', 'semantic_score', 'final_ranking_score']].sort_values(by='final_ranking_score', ascending=False).head(10))
```
